```python
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
Phenotypes = pd.read_csv('Phenotypes.csv')
Phenotypes.head()
```

```python
# For Alexander disease
HPOTerms = []
HPOFrequencies = []
for column in Phenotypes.columns:
    if 'HPO/HPOTerm' in column:
        HPOTerms.append(Phenotypes.loc[0,column])
    if 'HPOFrequency/Name/#text' in column:
        HPOFrequencies.append(Phenotypes.loc[0,column])
```

```python
AlexanderDisease = pd.DataFrame({
    'HPOTerms': HPOTerms,
    'HPOFrequencies': HPOFrequencies
})
AlexanderDisease.head()
```

```python
AlexanderDisease.to_csv('Associated_Phenotypes_With_AlexanderDisease.csv', index=False)
```
